# Which was the best month for sales and how much was old in that month

```python
all_month_df["Quantity Ordered"].dtype
all_month_df["Quantity Ordered"] = all_month_df["Quantity Ordered"].astype(int)
```

```python
all_month_df["Price Each"].dtype
all_month_df["Price Each"]=all_month_df["Price Each"].astype(float)
```

```python
all_month_df["Sale"] = all_month_df["Price Each"] * all_month_df["Quantity Ordered"]
```

```python
all_month_df["Sale"]
```

```python
sum_sales = all_month_df.groupby("Month")["Sale"].sum().sort_values(ascending = False)
sum_sales
```

 # Which city had the highest sales

```python
all_month_df["City"] = all_month_df["Purchase Address"].apply(lambda x : x.split(',')[1])
```

```python
all_month_df["City"]
```

```python
sale_per_city = all_month_df.groupby("City")["Sale"].sum().sort_values(ascending = False)
 
sale_per_city  = pd.DataFrame(sale_per_city) 
sale_per_city["Sale"] = sale_per_city["Sale"].apply(lambda s : '%.2f' % s )
sale_per_city["Sale"] = sale_per_city["Sale"].astype('float64')
sale_per_city = sale_per_city.reset_index()
sale_per_city
```

```python

plt.figure(figsize=(15,8))
sns.set(style ="dark", palette="colorblind", color_codes=True)

plt.bar(sale_per_city['City'],sale_per_city['Sale'] ,color=['black', 'red', 'green', 'blue', 'cyan'])

#sale_per_city['Sale'].plot(kind = "bar")
plt.xticks(sale_per_city['City'], rotation  = 'vertical', size = 10)
plt.xlabel("Cities")
plt.ylabel("Sales")
plt.title("Sales per city")
plt.show()
```
